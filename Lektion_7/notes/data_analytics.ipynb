```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

## Entpacken der Zip-Datei

```python
import zipfile

target = 'data/novel-corona-virus-2019-dataset.zip'
handle = zipfile.ZipFile(target)
handle.extractall('data/')
handle.close()
```

## Auslesen der Daten

```python
df = pd.read_csv('data/2019.csv')
df.head()
```
