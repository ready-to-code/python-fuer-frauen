# Visualisierung mit Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

## 1. Aufgabe
Gegeben sind folgendea Arrays:

```python
anteile = [32.9, 20.5, 10.7, 8.9, 9.2, 12.6, 5.0]
parteien = ["CDU/CSU", "SPD", "FDP", "Bündnis 90/Die Grünen", "Die Linke", "AfD", "Sonstige"]
```

Erstelle ein Tortendiagramm mit der `plt.pie()`-Methode, die die Bundestagswahlergebnisse 2017 repräsentieren.

```python
plt.pie(anteile,labels=parteien)
plt.title("Bundestagswahl 2017")
plt.show()
```

## 2. Aufgabe
Verwende das in Aufgabe 1 verwendete Beispiel und setze es in ein Balkendiagramm mit den gegebenen Farben um:

```python
farben = ['black', 'orangered', 'yellow', 'green', 'red', 'blue', 'grey']
```

```python
bars = plt.bar(parteien, anteile, color=farben) 
plt.title("Bundestagswahl 2017") 
plt.show()
```
